```python
import urllib.request
from bs4 import BeautifulSoup
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
```

```python
import pandas as pd
```

```python
county_district_mapping = {}
```

```python
for county_num in range(1, 255):
    url = "https://elections.sos.state.tx.us/elchist319_county" + str(county_num) + ".htm"
    
    headers={'User-Agent':user_agent,} 

    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    soup = BeautifulSoup(response)
    
    county = soup.find_all('h2')
    c = str(county[2]).split(">")[1].split("<")[0]
    print(c)
        
    all_tags = soup.find_all('td')
    all_tags = list(all_tags)
    
    districts = [str(x) for x in all_tags if "U. S. Representative" in str(x)]
    dn_list = []
    for d in districts:
        temp = d.split("District ")[1]
        dn = temp.split("<")[0]
        dn_list.append(dn)
        
    county_district_mapping[c] = dn_list
                                                                                                                                                    
```

```python
county_district_mapping
```

```python
output = pd.DataFrame.from_dict(county_district_mapping, orient='index')
```

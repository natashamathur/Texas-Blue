### Fixing problem 1

```python
x = pd.read_csv('final_data_for_model.csv')
```

```python
dc = pd.read_csv('main_district_for_counties.csv')
dc.columns = ['county', 'district']
dc = dc.reset_index()
```

```python
p = x[x.office=='president']
```

```python
dc_dict = {}
for idx, r in dc.iterrows():
    dc_dict[dc.iloc[idx].county] = dc.iloc[idx].district
```

```python
p.district = p.county.apply(lambda x: dc_dict[x])
```

```python
p.head()
```

```python
x = x.dropna(subset=['district'])
```

```python
x = x.append(p)
```
